# Linear Regression

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import KFold
```

```python
y = data_num['Balance']
y.shape
```

```python
X = data_num.drop(['Balance'], axis = 1)
X.shape
```

```python
model_linear = LinearRegression()
model_linear.fit(X,y)
R = model_linear.score(X,y)
R
```

```python
model_linear.coef_
```

```python
X_mat = np.array(X)
y_mat = np.array(y)
score = np.array([])
for j in np.arange(10):
    kf = KFold(n_splits = 3, shuffle = True, random_state = j)
    for train_index, test_index in kf.split(X_mat):
        mod_kf = LinearRegression()
        X_train = X_mat[train_index,] 
        y_train = y_mat[train_index]
        X_test = X_mat[test_index,] 
        y_test = y_mat[test_index]
    
        mod_kf.fit(X_train,y_train)
        score = np.append(score, mod_kf.score(X_test, y_test))
```

```python
score
```

```python
final_score = np.mean(score)
final_score
```

```python
R = final_score 
n = data_num.shape[0]
p = X.shape[1]
print(p)
1 - (1 - R)*(n-1)/(n-p-1)
```

# Ridge and Lasso

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LassoCV
from sklearn.linear_model import RidgeCV
```

```python
model_lasso = Lasso(normalize = True, alpha = 0.7)
model_lasso.fit(X,y)
model_lasso.score(X,y)
```

```python
col_list = X.columns
coef_list = model_lasso.coef_
list(zip(col_list, coef_list))
```

```python
model_lasso_cv = LassoCV(n_alphas = 500, cv = 5, max_iter = 10000, normalize = True)
model_lasso_cv.fit(X,y)
```

```python
model_lasso_cv.alpha_
```

```python
col_list = X.columns
coef_list = model_lasso_cv.coef_
list(zip(col_list, coef_list))
```

```python
model_lasso_cv.score(X,y)
```

```python
alphas_list = np.log10(model_lasso_cv.alphas_)

plt.figure(figsize = (12,7))
plt.plot(alphas_list, model_lasso_cv.mse_path_, ':')
plt.plot(alphas_list, model_lasso_cv.mse_path_.mean(axis=-1), 'k', label='Average across the folds', linewidth=2)
plt.axvline(np.log10(model_lasso_cv.alpha_), linestyle='--', color = 'k', label='alpha: CV estimate')
plt.legend()
plt.xlabel('log10(alpha)')
plt.ylabel('Mean square error')
plt.title('MSE on Each Fold')
plt.ylim(0,40000)
plt.xlim(-2,1)
```

```python
p
```

```python
alpha_list = model_lasso_cv.alphas_
coef_list = np.empty([p, len(alpha_list)])
j = 0
for alpha in alpha_list:
    mod = Lasso(alpha = alpha, max_iter = 10000, normalize = True)
    mod.fit(X,y)
    coef_list[:,j] = mod.coef_
    j = j + 1
```

```python
plt.figure(figsize = (15,7))
plt.plot(np.log10(alpha_list), coef_list.T)
plt.legend(data_num.columns)
plt.axvline(np.log10(model_lasso_cv.alpha_), linestyle='--', color = 'k')
plt.show()
```
